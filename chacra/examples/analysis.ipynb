```python
from chacra.ContactFrequencies import ContactFrequencies
```

### create the contact frequencies object

```python
cf = ContactFrequencies("./total_contacts.pd")
```

### you can view the contact frequencies as a pandas dataframe

```python
cf.freqs.loc[:5]
```

### and view the contact PCA loading scores 

```python
cf.cpca.loadings.iloc[:5]
```

### you can view the sorted normalized loadings by providing the chacra to sort on to the following method

```python
cf.cpca.sorted_norm_loadings(1)
```

```python
from chacra.plot import plot_chacras
import numpy as np
```

```python
fig = plot_chacras(cf.cpca, len(cf.cpca.top_chacras), cf.freqs, temps=np.geomspace(290,450,20), temp_scale='K')
```

The significance test claims there are 8 significant chacras. This only reflects 20 ns of sampling. More sampling will reduce the number of significant chacras and make these profiles appear smoother.

### 1tnf is a trimer. Let's average the data to see if the results look better

```python
from chacra.average import average_multimer
```

```python
avg = ContactFrequencies(average_multimer(structure = "./1tnf_truncated.pdb", df = cf.freqs))
```

```python
avg.cpca.sorted_norm_loadings(1)
```

```python
fig = plot_chacras(avg.cpca, len(avg.cpca.top_chacras), avg.freqs, temps=np.geomspace(290,450,20), temp_scale='K')
```

Notice that averagine the data between the 3 identical subunits effectively improves the sampling, reduces the number of significant chacras and makes the plots look smoother.

We don't have to take the significance test at face value to determine the number of chacras to focus on. With a couple hundred nanoseconds worth of sampling, this might be more useful. Let's look at the explained variance of the principal comoponents (chacras).

```python
from chacra.plot import plot_explained_variance
plot_explained_variance(avg.cpca)
```

After running for longer, the first principal component might account for 80% of the variance instead of ~50%. It looks like the first 3 PCs are interesting. 

```python
fig = plot_chacras(avg.cpca, 3, avg.freqs, temps=np.geomspace(290,450,20), temp_scale='K')
```

These are the main temperature response patterns of the protein's contacts. Let's look at where these contacts localize on the protein using the averaged data.

```python
from chacra.visualize.pymol import to_pymol
```

```python
contacts = []
for pc in range(1,4):
    # get the contacts with absolute normalized loading scores above 0.6 on the first 3 chacras
    contacts.extend(avg.cpca.get_chacra_center(pc,cutoff=0.6).index)
```

```python
to_pymol(contacts,
        avg.freqs,
        avg.cpca,
        output_file="1tnf_averaged_top_3_chacras.pml",
        pc_range=(1,3),
        variable_sphere_scale=True)
```
