```python
from chacra.ContactFrequencies import ContactFrequencies
```

### create the contact frequencies object

```python
cf = ContactFrequencies("./total_contacts.pd")
```

### you can view the contact as a pandas dataframe

```python
cf.freqs
```

### you can view the contact PCA data

```python
cf.cpca.loadings
```

### you can view the sorted normalized loadings by providing the chacra to sort on to the following method

```python
cf.cpca.sorted_norm_loadings(1)
```

```python
from chacra.plot import plot_chacras
import numpy as np
```

```python
fig = plot_chacras(cf.cpca, len(cf.cpca.top_chacras), cf.freqs, temps=np.geomspace(290,450,16))
```

### 1tnf is a trimer. Let's average the data to see if the results look better

```python
from chacra.average import average_multimer
```

```python
avg = ContactFrequencies(average_multimer(structure = "./1tnf_truncated.pdb", df = cf.freqs))
```

```python
avg.cpca.sorted_norm_loadings(1)
```

```python
fig = plot_chacras(avg.cpca, len(avg.cpca.top_chacras), avg.freqs, temps=np.geomspace(290,450,16))
```
